# Ejercicio 1

```

import numpy as np

# --- 1 y 2. CÁLCULO TEÓRICO ---
p_s1 = 0.4
p_s2 = 0.6

p_A_dado_s1 = (10*9*8*7*6) / (10**5)
p_A_dado_s2 = (10*9*8*7*6*5) / (10**6)

p_A = (p_A_dado_s1 * p_s1) + (p_A_dado_s2 * p_s2)
p_S1_dado_A = (p_A_dado_s1 * p_s1) / p_A

print(f"1. Teórica P(A) = {p_A:.5f}")
print(f"2. Teórica P(S1|A) = {p_S1_dado_A:.5f}")

# --- 3. SIMULACIÓN ---
N = 10000
casos_A = 0
casos_A_y_S1 = 0

np.random.seed(42)

for _ in range(N):
    sistema = np.random.choice(['S1', 'S2'], p=[p_s1, p_s2])

    if sistema == 'S1':
        codigo = np.random.randint(0, 10, 5)
    else:
        codigo = np.random.randint(0, 10, 6)

    if len(np.unique(codigo)) == len(codigo):
        casos_A += 1
        if sistema == 'S1':
            casos_A_y_S1 += 1

p_A_simulada = casos_A / N
p_S1_dado_A_simulada = casos_A_y_S1 / casos_A if casos_A > 0 else 0

print(f"3. Simulada P(A) = {p_A_simulada:.5f}")
print(f"3. Simulada P(S1|A) = {p_S1_dado_A_simulada:.5f}")
```

# Ejercicio 2

```

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import hypergeom

N = 12
K = 5
n = 4
rv = hypergeom(N, K, n)

k_vals = np.arange(0, 5)

probabilidades = rv.pmf(k_vals)
print("1. Probabilidades P(X=k):")
for k, p in zip(k_vals, probabilidades):
    print(f"P(X={k}) = {p:.4f}")

distribucion = rv.cdf(k_vals)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.vlines(k_vals, 0, probabilidades)
ax1.plot(k_vals, probabilidades, 'o')
ax1.set_title('f_X')

ax2.step(k_vals, distribucion, where='post')
ax2.set_title('F_X')

plt.show()

print("Media:", rv.mean())
print("Varianza:", rv.var())
print("Desviación típica:", rv.std())

simulaciones = rv.rvs(size=10000, random_state=42)
print("Media simulada:", np.mean(simulaciones))
print("Varianza simulada:", np.var(simulaciones))
```

# Ejercicio 3

```

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

fichero = 'dat1.txt'
datos = np.loadtxt(fichero)

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)

es_discreto = np.all(np.mod(datos, 1) == 0)

if es_discreto:
    bins = np.arange(datos.min()-0.5, datos.max()+1.5, 1)
    plt.hist(datos, bins=bins, density=True)
else:
    plt.hist(datos, bins=20, density=True)

plt.title(f'Histograma de {fichero}')

media = np.mean(datos)
std = np.std(datos, ddof=1)

if es_discreto:
    dist = stats.poisson(mu=media)
else:
    dist = stats.norm(loc=media, scale=std)

plt.subplot(1, 2, 2)
deciles = np.arange(0.1, 1.0, 0.1)

qm = np.quantile(datos, deciles)
qt = dist.ppf(deciles)

plt.scatter(qt, qm)
plt.plot([min(qt), max(qt)], [min(qt), max(qt)])

plt.title('Q-Q plot')
plt.show()
```
